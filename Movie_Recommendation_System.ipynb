```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("movies (1).csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
# selecting features for filling the null values
sel_features = ['genres','keywords','popularity','tagline','cast','director']
```

```python
data.isnull().sum()
```

```python
# filling up the null values with blank spaces
for feature in sel_features:
    data[feature] = data[feature].fillna("")
```

```python
data.isnull().sum()
```

```python
# Combining all the selected features
comb_features = data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
```

```python
comb_features
```

```python
# converting the text data into the vectors
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vect = TfidfVectorizer()
```

```python
# converting the text data into the vectors
feature_vector = vect.fit_transform(comb_features)
```

```python
feature_vector
```

```python
# Importing the cosine similarity for similarity score
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(feature_vector)
```

```python
similarity.shape
```

```python
# Taking the Movie Name input from the user
movie_name = input('Enter the movie name in which are interested : ')
```

```python
# making a movies list for difflib
movies_list = data['title'].tolist()
```

```python
# importing difflib for the most similar movie title compared to user input 
import difflib
# Find close matches
close_matches = difflib.get_close_matches(movie_name, movies_list)
close_matches
```

```python
close_match = close_matches[0]
close_match
```

```python
# finding the index of the movie title
movie_index = data[data.title == close_match]['index'].values[0]
movie_index
```

```python
# checking the similarity score of other titles compared to the user input one
similarity_score = list(enumerate(similarity[movie_index]))
similarity_score
```

```python
len(similarity_score)
```

```python
# sorting the movies based on the similarity score in descending order
sorted_similarity_score = sorted(similarity_score, key= lambda x:x[1], reverse = True)
sorted_similarity_score
```

```python
# Find the name of the similar movies based on the index

i = 1
for movie in sorted_similarity_score:
    index = movie[0]
    title_of_the_movie = data[data.index == index]['title'].values[0]
    
    if (i<11):
        print(i, title_of_the_movie)
        i+=1
```
